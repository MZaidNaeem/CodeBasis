```python
import numpy as np

def gradient_descent(x, y, lr=0.01, epocs=10000):
    m = 0.0
    b = 0.0
    
    for epoch in range(epocs):
        y_p = m * x + b
        error = y - y_p

        cost = np.mean(error**2)

        dm = -2 * np.mean(error * x)
        db = -2 * np.mean(error)

        m = m - lr * dm
        b = b - lr * db

        print(f"Epoch {epoch}: Cost = {cost:.4f}, m = {m:.4f}, b = {b:.4f}, dm = {dm * lr:.4f}, db = {db * lr:.4f}")

if __name__ == "__main__":
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([5, 7, 9, 11, 13]) 
    gradient_descent(x, y)
```
