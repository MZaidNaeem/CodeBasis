```python
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import seaborn as sns

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
)

X = StandardScaler().fit_transform(X)

df = pd.DataFrame({"f1": X[:, 0], "f2": X[:, 1]})
df.head()
```

```python
sns.scatterplot(df, x="f1", y='f2')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

ds = DBSCAN(eps=0.3, min_samples=10)
ds.fit(df[["f1", "f2"]])
df['cluster'] = ds.labels_
df.head()
```

```python
sns.scatterplot(df, x="f1", y='f2',hue='cluster')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_excel("income.xlsx", names=["name","age","income"])
df
```

```python
sns.scatterplot(df, x='age', y='income', s=100)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

ds = DBSCAN(eps=10, min_samples=2)
ds.fit(df[['age','income']])
df['cluster'] = ds.labels_
df.head()
```

```python
sns.scatterplot(df, x='age', y='income', s=100,hue='cluster')
plt.show()
```
