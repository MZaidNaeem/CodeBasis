```python
# For numerical computations
import numpy as np

# For reading CSV or Excel files
import pandas as pd

# For plotting (scatter plots, regression lines, etc.)
import matplotlib.pyplot as plt

# For basic linear regression and polynomial feature transformation
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

```

```python
df=pd.read_csv("house_price_regression_dataset.csv")
print(df.shape)
df.head()
df.replace(["yes","no","furnished","semi-furnished","unfurnished"],[1,0,2,1,0],inplace=True)
print(df.isna().sum())
print((df.dtypes=="int64"))
df
```

```python
X=df.drop(['House_Price'],axis=1)
y=df['House_Price']

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
degree=2;
poly=PolynomialFeatures(degree);
train_poly=poly.fit_transform(x_train);
test_poly=poly.transform(x_test)
```

```python
model=LinearRegression();
model.fit(train_poly,y_train);
r=model.score(test_poly,y_test);
r
```

```python
model1=Ridge(alpha=20);
model1.fit(train_poly,y_train);
r=model1.score(test_poly,y_test);
r
```
