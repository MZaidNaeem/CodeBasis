```python
import pandas as pd
import numpy as np

# Assuming you have already read the data into 'df'
# Example: df = pd.read_csv('your_file.csv', header=None)

# Find the indices where each page starts
page_starts = df[df.iloc[:, 0].str.contains("Page", na=False)].index.tolist()
page_starts.append(len(df))  # Add the end of the DataFrame for slicing

# Separate the data for each page
df_page1 = df.iloc[page_starts[0] + 1:page_starts[1], 1:].reset_index(drop=True)
df_page2 = df.iloc[page_starts[1] + 1:page_starts[2], 1:].reset_index(drop=True)
df_page3 = df.iloc[page_starts[2] + 1:page_starts[3], 1:].reset_index(drop=True)

# Set appropriate column names
df_page1.columns = ['ColumnA', 'ColumnB', 'ColumnC', 'ColumnD']
df_page2.columns = ['ColumnE', 'ColumnF', 'ColumnG', 'ColumnH']
df_page3.columns = ['ColumnI', 'ColumnJ', 'ColumnK', 'ColumnL']

print("DataFrame for Page 1:")
print(df_page1, "\n")

print("DataFrame for Page 2:")
print(df_page2, "\n")

print("DataFrame for Page 3:")
print(df_page3)
```
