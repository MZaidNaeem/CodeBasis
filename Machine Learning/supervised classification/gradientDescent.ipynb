```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (you can replace this with your own)
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 6, 8, 10])  # y = 2x (perfect line)

# Initialize parameters
m = 0  # slope
b = 0  # intercept

# Hyperparameters
learning_rate = 0.01
epochs = 1000
n = len(X)

# Gradient Descent
for i in range(epochs):
    y_pred = m * X + b
    error = y_pred - Y

    cost = (1/n) * sum(error ** 2)
    
    # Partial derivatives
    dm = (2/n) * sum(error * X)
    db = (2/n) * sum(error)

    # Update parameters
    m -= learning_rate * dm
    b -= learning_rate * db

    
    print(f"Epoch {i}: m = {m:.4f}, b = {b:.4f},dm={dm}, db={db}, cost = {cost:.4f}")

print(f"\nFinal equation: y = {m:.4f}x + {b:.4f}")

# Plotting the result
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, m*X + b, color='red', label='Best Fit Line')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gradient Descent for Linear Regression')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Mean Squared Error Loss
def mse_loss(y, y_hat):
    return (y - y_hat) ** 2

# Binary Cross-Entropy Loss
def bce_loss(y, y_hat):
    return - (y * np.log(y_hat + 1e-8) + (1 - y) * np.log(1 - y_hat + 1e-8))

# Generate a range of z values (logits)
z = np.linspace(-10, 10, 200)
y_hat = sigmoid(z)

# True label: let's assume y = 1 (positive class)
y_true = 1

# Calculate both losses
mse = mse_loss(y_true, y_hat)
bce = bce_loss(y_true, y_hat)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(z, mse, label='MSE Loss', color='blue')
plt.plot(z, bce, label='Binary Cross-Entropy Loss', color='red')
plt.title('MSE vs Cross-Entropy Loss (y = 1)')
plt.xlabel('Logit (z)')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
