```python
f = pn.read_excel("mymovies.xlsx")
f
```

```python
ff=f.iloc[:10,:]
fe=f.iloc[10:,:]

ff.reset_index(inplace=True)
ff=ff.drop(columns=["index"])
ff=ff.drop(index=0)

fe.reset_index(inplace=True)
fe=fe.drop(columns=["index"])
fe=fe.drop(index=0)

with pn.ExcelWriter("final.xlsx") as writer:
    ff.to_excel(writer, sheet_name="data1", index=False)
    fe.to_excel(writer, sheet_name="data2", index=False)

```
